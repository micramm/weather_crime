```python
import pandas as pd
import datetime
```

```python
ser_rain = pd.Series.from_csv('working_data/SF_rainfall.csv')
df_crime = pd.read_csv('working_data//sfpd_incident_2013.csv')
```

```python
#convert this to a proper datetime expression
def to_datetime(date_string):
    year,month,day = date_string.split('-')
    return datetime.datetime(int(year),int(month),int(day))

df_crime['date'] = df_crime.date.apply(to_datetime)
```

```python
crime = df_crime.groupby('date').incidntnum.count()
```

```python
together = pd.DataFrame({'crime':crime, 'rain':ser_rain})
together['rain'] = together.rain > 0
together = together.dropna()
```

```python
def calc_reduction(column = 'rain'):
    grouped = together.groupby(column).crime.mean()
    reduction = grouped[True] / grouped[False]
    return 1 - reduction

reduction = calc_reduction()
print reduction
```
